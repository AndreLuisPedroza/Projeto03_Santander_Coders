## **Perguntas sobre o Dataset utilizando SQL**

1 - Qual sabor de pizza teve a maior demanda de vendas?

```python
ps.sqldf("""
SELECT
    pt.name AS "Nome_do_Sabor",
    SUM(od.quantity) AS "Quantidade_total"
FROM orders_details AS od
LEFT JOIN pizzas AS p ON od.pizza_id = p.pizza_id
LEFT JOIN pizza_type AS pt ON p.pizza_type_id = pt.pizza_type_id
GROUP BY p.pizza_type_id
ORDER BY Quantidade_total DESC
LIMIT 3
""")
```

2 - Qual sabor de pizza teve a menor demanda de vendas?

```python
ps.sqldf("""
SELECT
    pt.name AS "Nome_do_Sabor",
    SUM(od.quantity) AS "Quantidade_total"
FROM orders_details AS od
LEFT JOIN pizzas AS p ON od.pizza_id = p.pizza_id
LEFT JOIN pizza_type AS pt ON p.pizza_type_id = pt.pizza_type_id
GROUP BY p.pizza_type_id
ORDER BY Quantidade_total ASC
LIMIT 3
""")
```

3 - Em quais meses ocorreram as maiores vendas de pizzas?

```python
ps.sqldf("""
SELECT
    o.month AS "Mês",
    SUM(od.quantity) AS "quantidade_total"
FROM orders_details AS od
LEFT JOIN orders AS o ON o.order_id = od.order_id
GROUP BY o.month
ORDER BY quantidade_total DESC
LIMIT 3
""")
```

4 - Em quais meses ocorreram as menores vendas de pizzas?

```python
ps.sqldf("""
SELECT
    o.month AS "Mês",
    SUM(od.quantity) AS "quantidade_total"
FROM orders_details AS od
LEFT JOIN orders AS o ON o.order_id = od.order_id
GROUP BY o.month
ORDER BY quantidade_total ASC
LIMIT 3
""")
```

5 - Quais foram os três meses com a receita mais alta?

```python
ps.sqldf("""
SELECT
    o.month AS "mes",
    SUM(p.price) AS "montante"
FROM orders_details AS od
LEFT JOIN pizzas AS p ON p.pizza_id = od.pizza_id
LEFT JOIN orders AS o ON o.order_id = od.order_id
GROUP BY o.month
ORDER BY montante DESC
LIMIT 3
""")
```

6 - Quais foram os três meses com a receita mais baixa?







```python
ps.sqldf("""
SELECT
    o.month AS "mes",
    SUM(p.price) AS "montante"
FROM orders_details AS od
LEFT JOIN pizzas AS p ON p.pizza_id = od.pizza_id
LEFT JOIN orders AS o ON o.order_id = od.order_id
GROUP BY o.month
ORDER BY montante ASC
LIMIT 3
""")
```

7 - Qual foi o total de vendas para cada categoria de pizza?

```python
ps.sqldf("""
SELECT
    pt.category AS "categoria",
    SUM(od.quantity) AS "quantidade_total_vendas"
FROM orders_details AS od
LEFT JOIN pizzas AS p ON p.pizza_id = od.pizza_id
LEFT JOIN pizza_type AS pt ON pt.pizza_type_id = p.pizza_type_id
GROUP BY pt.category
ORDER BY quantidade_total_vendas DESC
""")
```

9 - Qual a média de preço por tamanho de pizza da pizzaria?

```python
ps.sqldf("""
SELECT
    size AS "tamanho_pizza",
    AVG(price) AS "media"
FROM pizzas
GROUP BY size
""")
```

10 - Quantos tipos diferentes de sabores de pizza existem em cada categoria?

```python
ps.sqldf("""
SELECT
     category AS "Categoria",
     COUNT(category) AS "Total_Categoria"
FROM pizza_type
GROUP BY category
""")
```

11 - Qual a quantidade de vendas por tamanho de pizza?

```python
ps.sqldf("""
SELECT
    p.size AS "tamanho",
    SUM(quantity) AS "total_vendas"
FROM orders_details AS od
LEFT JOIN pizzas AS p ON p.pizza_id = od.pizza_id

GROUP BY p.size
""")
```

12 - Considerando o tamanho 'L', quais são os sabores de pizza mais caros do cardápio?

```python
ps.sqldf("""
SELECT
    pt.name AS "sabor_da_pizza",
    p.price AS "preco"
FROM pizzas AS p
LEFT JOIN pizza_type AS pt ON pt.pizza_type_id = p.pizza_type_id
WHERE p.size = 'L'
ORDER BY p.price DESC
LIMIT 3
""")
```
